```r
data <- read.csv('data_with_titles.csv', header=FALSE, sep=";")

# Add column headers
names(data) <- c('RePEc Handle', 'Year', 'Citations', 'IF', 'IF 5 Years', 'Documents_Published', 'H_Index', 'Type', 
                 'ISSN 1', 'ISSN 2', 'Type 2', 'Year 3', 'Scimago Source ID', 'SJR', 'Type 4', 'Year 5', 'Jourqual Category', 
                 'Type 6', 'Year 7', 'BWL', 'Type 8', 'Year 9', 'VWL', 'Title')

# Remove NA values
data[is.na(data)] <- 0
```

```r
plot(data$Citations, data$Documents_Published)
```

```r
plot(data)
```

```r
plot(data$Citations, data$Documents_Published, type="b")  # points & lines
plot(data$Citations, data$Documents_Published, type="h")  # high density vertical lines
plot(data$Citations, data$Documents_Published, type="c")
plot(data$Citations, data$Documents_Published, type="s")
```

```r
plot(data$Citations, data$Documents_Published, 
     xlab='Citations', ylab='Publications', 
     main="Citations and Publications", col="green")
```

```r
barplot(data$Citations,
        main='Citations',
        xlab='Citations',
        horiz='TRUE',
        col='green')
```
